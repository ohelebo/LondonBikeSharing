```python
# CONVERT THESE VALUES TO ACTUAL STRINGS AND BOOLEAN (for WEEKEND)
```

```python
weather_val = {
    '1.0': 'Clear Skies',
    '2.0': 'Scattered Clouds',
    '3.0': 'Broken Clouds',
    '4.0': 'Cloudy',
    '7.0': 'Showers',
    '10.0':'Rain with Thunderstorm',
    '26.0': 'Snowfall',
    '94.0': 'Freezing Fog'
}
```

```python
# Since it's a yes or no (boolean) kind of response, this variable will be used
#for both the holiday and weekend columns
hol_wknd_val = {
    '0.0': 'No',
    '1.0': 'Yes'
    }
```

```python
seasn_val = {
    '0.0': 'Spring',
    '1.0': 'Summer',
    '2.0': 'Autumn',
    '3.0': 'Winter'
}
```

```python
#Convert the data types in the Weather, Holiday, Weekend and Season Columns to Strings
#Then map the already created dic changes to it
```

```python
data.Weather = data.Weather.astype('str')
data.Weather = data.Weather.map(weather_val)
```

```python
data.Holiday =  data.Holiday.astype('str')
data.Holiday = data.Holiday.map(hol_wknd_val)
```

```python
data.Weekend = data.Weekend.astype('str')
data.Weekend = data.Weekend.map(hol_wknd_val)
```

```python
data.Season = data.Season.astype('str')
data.Season = data.Season.map(seasn_val)
```

```python
#Convert data these columns into categorical data so it can be used in Power BI as a slicer, create visualizations
#that leverage the categorical nature of the data, and enable features like drill-down or hierarchy.
```

```python
data_to_cat = ['Weather', 'Holiday', 'Weekend', 'Season']
```

```python
for column in data_to_cat:
    data[column] = data[column].astype('category')
```
